## Hypothesis 3: The mean budget for movies which passes all tests is lesser than the movies which don't pass

```python
budget=movie_dataset.groupby('clean_test')['budget'].mean()
budget=pd.to_numeric(budget,downcast='float')
budget
```

## Answer: The budget is less for movies which pass all the three tests

```python
import matplotlib.pyplot as plt1
budget.plot.bar()
plt1.xlabel('categories')
plt1.ylabel('budget in thousands')
plt1.title('Budget')
```

```python
international_grossing=movie_dataset.groupby('clean_test')['intgross'].mean()
international_grossing=pd.to_numeric(international_grossing,downcast='float')
international_grossing
```

```python
domestic_grossing=movie_dataset.groupby('clean_test')['domgross'].mean()
domestic_grossing=pd.to_numeric(domestic_grossing,downcast='float')
domestic_grossing
```

```python
df = pd.DataFrame({'International Grossing': international_grossing,'Domestic Grossing': domestic_grossing,'Budget': budget})
df.plot.bar()
plt1.title('Grossing vs Budget')
plt1.xlabel('categories')
plt1.ylabel('Dollars in millions')
```
